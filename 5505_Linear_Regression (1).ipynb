##Multivariate model without any changes - model 1

```python
x = data[['HEIGHT', 'WIDTH', 'SIGNED', 'PICTURE', 'HOUSE ']]

y = data[["PRICE"]]
```

```python
##Without changing any data
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=101)
```

```python

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train, y_train)

predictions = lm.predict(x_test)

from sklearn.metrics import mean_squared_error
print("RMSE of the linear regression model :" +str(np.sqrt(mean_squared_error(predictions,y_test))))
```

##Root mean square error is 3.55

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train, y_train)

predictions = lm.predict(x_test)

from sklearn.metrics import mean_squared_error
print("RMSE of the linear regression model :" +str(np.sqrt(mean_squared_error(predictions,y_test))))
```

## There is a significant drop in the RMSE value after performing standardization on the output variable and removing HEIGHT due to high corelation 

## Multivariate LR model 4

```python
##Normalizing the data
#ie. using minmax scalar from sklearn
```

```python
nom_data=pd.read_csv("monet.csv")
```

```python
x = nom_data[['HEIGHT', 'WIDTH', 'SIGNED', 'PICTURE', 'HOUSE ']]

y = nom_data[["PRICE"]]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python

n = MinMaxScaler().fit(x_train)

x_train_norm = n.transform(x_train)

x_test_norm = n.transform(x_test)

```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train_norm, y_train)

predictions = lm.predict(x_test_norm)

from sklearn.metrics import mean_squared_error
print("RMSE of the linear regression model :" +str(np.sqrt(mean_squared_error(predictions,y_test))))
```

##Makes no difference same as original
