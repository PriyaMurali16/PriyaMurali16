```python
quality_classified=[]
for i in wine["quality"]:
 if i>=6:
    quality_classified.append(1);
 else:
    quality_classified.append(0);
 i=+1
```

```python
quality_classified
```

```python
wine["quality_c"]=quality_classified
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=wine[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
y=wine["quality_c"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=101,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

> without any changes

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predicted = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_test, predicted) 
```

```python
print("Classification Report: ", "\n", "\n",report)
```

```python
#score the accuracy level

result = model.score(x_test, y_test)

#print out the results

print(("Accuracy: %.3f%%") % (result*100.0))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, predicted)
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(mat, annot=True)
```

##Standardizing the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_sc=StandardScaler()
x_train_ss=std_sc.fit_transform(x_train)
x_test_ss=std_sc.transform(x_test)
```

```python
model.fit(x_train_ss,y_train)
```

```python
predicted=model.predict(x_test_ss)
```

```python
#score the accuracy level

result = model.score(x_test_ss, y_test)

#print out the results

print(("Accuracy: %.3f%%") % (result*100.0))
```
